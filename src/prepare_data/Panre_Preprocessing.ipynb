# Panre Preprocessing

```python
import re
import pandas as pd
```

```python
panre_text_df = pd.read_csv('panre_text.csv')
```

```python
def clean_html(text):
    if not isinstance(text, str):
        return text
    # <br/> 태그를 \n으로 변환
    text = re.sub(r'<br\s*/?>', '\n', text)
    # 나머지 태그 제거
    text = re.sub(r'<[^>]+>', '', text)
    # 중복 개행 문자를 하나로 줄임
    text = re.sub(r'\n+', '\n', text)
    
    return text
```

```python
# 테스트
html_text = "이것은 예시 텍스트입니다.<br/><br/>새로운 줄입니다.<div>이 부분은 삭제됩니다.</div><br/>끝입니다."

# 함수 호출 및 결과 출력
cleaned_text = clean_html(html_text)
print(cleaned_text)
```

```python
import numpy as np

html_text_2 = np.nan

cleaned_text_2 = clean_html(html_text_2)
print(type(cleaned_text_2))
```

```python
panre_text_df['판시사항'] = panre_text_df['판시사항'].apply(clean_html)
panre_text_df['판결요지'] = panre_text_df['판결요지'].apply(clean_html)
panre_text_df['참조조문'] = panre_text_df['참조조문'].apply(clean_html)
panre_text_df['참조판례'] = panre_text_df['참조판례'].apply(clean_html)
panre_text_df['판례내용'] = panre_text_df['판례내용'].apply(clean_html)
```
